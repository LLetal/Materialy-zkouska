# Práce a pV diagram
Práce vykonaná při změně objemu (odvozeno již v kapitole o kinetické teorii plynů):
$$
dW = pdV
$$
Použijeme stavovou rovnici ideálního plynu ve tvaru:
$$
pV  = NkT => W = \int_{V_1}^{V_2}\frac{NkT}{V} dV = NkT ln(\frac{V_2}{V_1})
$$
Při rovnovážném stavu se rovná práce vykonaná na soustavu s prací, kterou vykoná soustava

# Termodynamické zákony
Poprvé odvodil nějaký angličan pomocí velmi podrobné logické analýzy.
## 1. Termodynamický zákon
Teplo dodané systému v součtu prací vykonanou na systému zvyšují celkovou energii systému U. (Opisně řečený zákon zachování energie.) Přímo z ní potom plyne i druhá formulace, a sice, že teplo dodané systému je rovno přírůstku energie systému a prací systémem vykonanou. Matematicky:
$$
Q + W = \Delta U
$$

$$
Q = \Delta U + W'
$$
V izolované soustavě se energie nemění- > zůstává konstantní -> ZZE.
Kruhový děj = Všechno teplo se přemění na práci vykonanou systémem a nijak se nenavýší vnitřní energie
## 2. Termodynamický zákon
Teplo nemůže samovolně přecházet z chladnějšího místa/tělesa na teplejší. Z pohledu kinetické teorie bychom museli při srážkách částic nějakým způsobem vynutit srážení ve prospěch rychlejších částic, což se prokazatelně nemůže stát (ZZH, ZZE).
## 3. Termodynamický zákon (Nerstův)
Při absolutní nule je entropie nulová -> Entropie vždy roste

## Entropie
Je to stavová veličina **S(V,T)** Já si jí interpretuji jako míru všech možných stavů, kterých můžou objekty uvnitř systému v následujím okamžiku. Při vratných procesech se zachovává (nepřidáváme do systému žádný nový stav) a při nevratných vždy narůstá. 
$$
dS = \frac{dQ}{T}
$$
Změna entropie mezi dvěma stavy $A$ a $B$ je rovna:
$$
\Delta S = S(V_B,T_B) - S(V_A, T_B) = \int_{A}^{B} \frac{dQ}{T}
$$

# Carnotův cyklus
Právě takový stroj, který je dokonale vratný (v tom smyslu, že můžeme měnit směr vykonávání úkonů - od komprese k expanzi). Pro práci vykonanou pomocí Carnotova stroje platí, že jakékoliv jiné stroje vždy vykonají při své činnosti menší práci, než by vykonal Carnotův.
Jednotlivé kroky cyklu:
### Izotermická expanze
Veškeré teplo dodávané podložkou přeměňujeme na práci tak, že velmi pomalu zvyšujeme objem nádoby. (práci koná podložka -> všechno dodané teplo plynu se spotřebuje na expanzi, Teplota plynu uvnitř válce je $T_1$)
### Adiabatická expanze
V této fázi odstraňujeme válec z podložky a přesouváme jej mezi obě nahřívací podložky. Posouváme pomalu pístem, dokud teplota uvnitř válce neklesne na teplotu druhé (chladnější) podložky. Důvod, proč plyn uvnitř válce chladne je ten, že spotřebováváme jeho vnitřní energii na expanzi
### Izotermiké stlačení plynu
Přiložíme váleček k podložce o teplotě $T_2$ a začneme jej stlačovat, čímž konáme práci, která se spotřebovává na teplo. Teplota plynu nevzroste -> kvůli kontaktu s podložkou (Stejně jako při izotermické expanze). 
### Adiabatická komprese
Jakmile odevzdáme druhé podložce nějaké teplo (snížíme vnitřní energii plynu uvnitř válce) odebereme jej z podložky a necháme v prostoru mezi podložkami. Plyn poté stlačujeme tak dlouho, dokud opět nezíská tepltou $T_1$, čímž se ale vrátíme do původního stavu.


# Účinnost Carnotova stroje
Protože víme, že vykonaná práce je rovna rozdílu tepla přijatého při izotermické expanzi a tepla odevzdaného při kompresi. Začněme tím, že si definujeme, jaké teplo získá při expanzi. Protože vnitřní aenergie systému se nemění (teplota i počet částic je stejný), bude platit:
$$
W = \int_{V_b}^{V_b} p dV
$$
Z rovnice ideálního plynu poté platí $p = \frac{NkT_1}{V}$:
$$
Q_1 = \int_{V_a}^{V_b} \frac{NkT_1}{V} dV = NkT ln(\frac{V_b}{V_a})
$$
(Stejné jako na počátku). Pro teplo odevzdané při kompresi potom bude platit vztah získaný identickým postupem:
$$
Q_2 = NkT_2 ln (\frac{V_c}{V_d})
$$
Nyní potřebujeme nalézt, jaký je vztah mezi mezi poměry objemů $\frac{V_c}{V_d}$ a $\frac{V_a}{V_b}$. Ten získáme z částí adiabatiké komprese a expanze, kdy víme, že musím platit, během nich musí $pV^\gamma = konst$ v obou případech. Můžeme tedy napsat:
$$
\frac{V_b}{V_a} = \frac{V_c}{V_d} 
$$
A konečně
$$
\eta = Q_1 - Q_2 = Q_1 (1-\frac{T_2}{T_1})
$$
## Termodynamická definice teploty
Teplota tělesa = kolik tepla absorboval vratný stroj pracující mezi teplotou tělesa a jednotkovou teplotou (nezávisí na materiálu v kterém operuje)
