# Vlnová funkce
Nejprve je dobré si říci, že šíření vlny v prostoru bude s jistotou popisovat nějaká funkce tvaru $f(x-ct)$, neboť právě tento tvar zaručuje, že se po určité době bude opakovat. Stěžejní funkce, ze které budeme vycházet je funkce popisující polohu vlny v závislosti na čase $\gamma(x,t)$. Současně za zmínku stojí, že aby mohl zvuk vzniknout, musí být střední volná dráha mnohem menší, než vzdálennost, na které se zvuk šíří - jinak by nemohlo docházet ke srážkám a šíření vzruchu.
Odvození vlnové funkce na příkladu zvuku se vychází z tří skutečností:
 - I. Plyn se pohybuje a mění se tedy jeho hustota
 - II. Změně hustoty odpovídá změna tlaku
 - III. Nerovnoměrné rozdělení tlaku vyvolává pohyb plynu \
Něco jako důkaz: \

### Druhá část

Když se plyn pohybuje, tak se mění jeho hustota a s ní i tlak. Máme tedy jistou funkci:
$$
p = f(\rho)
$$
Víme, že dodatečný tlak, který způsobuje pro nás slyšitelný vzduch je ve srovnání s rovnovážným tlakem velmi malý. Vztah mezi tlakem a hustotou tedy je:
$$
p_0 + p_e = f(\rho_0 + \rho_e) =  f(\rho_{0}) + \rho_{e}f'(\rho_0)
$$
Neboť každému tlaku přísluší nějaká hustota. Protože je ale změna té hustoty velmi malá, můžeme zapsat. Z tohoto vztahu rovněž plyne, že nový tlak je přímo úměrný vyvolané hustotě o nějaký koeficient $\kappa$, tedy $p_e=\rho_e \kappa$. Tímto jsme odvodili právě druhý bod závislost a můžeme se přesunout na první.
### První část
Je-li posunutí vlny v čase t způsobené zvukovou vlnou $\gamma(x,t)$, tak začíná-li se vlna šířit z bodu x a konec "kapsy" je v  poloze $x +\Delta$, nová poloha potom vlivem šíření je $x+gamma(x,t)$$x+\Delta x +gamma(x+\Delta,t)$ apod... (viz obrázek.)

Předpokládáme-li, že se se zachovává množství částic v objemu (mělo by), tak ve starém objemu bude zajisté platit, že:

$$
m = V \rho_0 = \rho_0 \Delta
$$
(předpokládáme jednotkovou plochu kolmou na pohyb vlny, totéž v druhém případě). Poté ale musí notně platit, že bude stejné množství částic i v objemu $V = 1 (x+\Delta x + \gamma(x+\Delta x,t) -x - \gamma(x,t))$, přestože dojde ke změně hustoty na $\rho$. Bude tedy platit rovnost:
$$
\rho_0 \Delta x = \rho (\Delta x + \gamma(x+\Delta x,t) - \gamma(x,t))
$$
Povšimněme si, že člen $\gamma(x+\Delta x,t) - \gamma(x,t)$ je definicí parciální derivace funkce gamma podle x (předpokládáme, že $\Delta x$ je nesmírně malé). Potom za předpokladu, že nová hustota je rovna $\rho = \rho_0 + \rho_e$(Už v předchozí části jsme si to ověřili) získáváme:
$$
\rho_0 \Delta x = (\rho_0 + \rho_e)(\frac{\partial \gamma}{\partial x}\Delta x + \Delta x)
$$
Roznásobením získáme a vykrácením $\Delta x$ získáme:
$$
\rho_e = -\rho_0 \frac{\partial \gamma}{\partial x} - \rho_e \frac{\partial \gamma}{\partial x}
$$
Znovu využijeme skutečnosti, že počítáme s hrozně malými věcmi, pročež druhý člen vypustíme, čímž získáme:
$$
\rho_e = \rho_0 \frac{\partial \gamma}{\partial x}
$$
Znaménko - je zde proto, že s rostoucím posunutím se hustota přeci snižuje, jak se vlna šíří prostorem.
### III. vztah
Zbývá nám najít poslední rovnici. Vezmeme-li opět nějaký drobný objem vzduchu o délce $\Delta x$ a ploše o jednotkové velikosti, potom hmotnost tohoto objemu bude:
$$
m = \rho_0 \delta x
$$
Síla působící na tento malý segment vlny bude mít z 2. newtonova zákona velikost:
$$
F = \rho_0 \Delta x \frac{\partial^2 \gama}{\partial t^2}
$$
V bodě x na objem však působit síla rovná:
$$
F(x) = p(x,t) 
$$
a proti konci vlny potom nějaká další síla:
$$
F(x+ \Delta x) = p(\Delta x +x,t)
$$
(Člen obsahu vypadává, protože je jednotkový. ). Po sečtení sil opět zjišťujeme, že jsme nalezli parciální derivace a 2. Newtonův zákon má tvar:
$$
\rho_0 \frac{\partial^2 \gama}{\partial t^2} = \frac{\partial\gamma}{\partial x} p_e
$$

(funkce $p(x,t) $ se pochopitelně skládá z dvou částí $p_0 $ a $p_e$, nicméně první člen nám vypadává, neboť je konstantní a parciální derivace jej tedy smaže.).
### Zkombinování všech členů
Zkombinujeme-li nyní všechny podúkoly dohromady, získáme rovnici:
$$
\frac{\partial^2 \gamma(x,t)}{\partial t^2} \frac{1}{\kappa} = \frac{\partial^2 \gamma(x,t)}{\partial x^2}
$$
Ihned na počátku jsme si ukázali, že funkce popisující pohyb vlny musí notně mít tvar $f(x-vt)$, pročež se pokusíme do rovnice tuto funkci dosadit, abychom zjistili, zda vyhovuje. CO by to bylo za rovnici funkce, kdyby neodpovídala chování vlny.
$$
\rho_0 \frac{\partial^2 f(x-vt)}{\partial t^2} = \frac{\partial^2 f(x-vt)}{\partial x^2} 
$$ 

Když diferencujeme funkci podle x, tak získáme pouze $f''(x-vt)$, pakliže podle t, získáváme však $v^2 f''(x-vt)$, čili:
$$
f''(x-vt) = \frac{1}{\kappa} v^2 f(x-vt)
$$

Odsud navíc plyne, že $\kappa = v^2$, aby rovnost platila. Snadno ověříme, že vlnová funkce platí i pro složené vlmny a stejně tak i pro vlnění pohybující se opačným směrem.

## Odvození rychlosti zvuku
UVnitř vlny je mezi tlakem a teplotou adiabatická závislost. Platí tedy:
$$
pV^{\kappa} = konstanta
$$
Pakliže je ale hustota ve vztahu k objemu v nepřímé úměře, tak pro tlak platí:
$$
p = konstanta \rho ^\kappa
$$
V minulé části jsem si odvodili, že pro rychlost v prostředí platí vztah $v = \frac{\partial p}{\partial \rho}$, čili po zderivování dostáváme:
$$
v^2 = \frac{\kappa p}{\rho}
$$
Využijeme-li stavové rovnice ideálního plynu $pV = nRT$, tak můžeme z rovnice dosadit za tlak a dále upravovat, čímž získáme:
$$
v^2 = \frac{\kappa nRT}{\rho V} = \frac{\kappa nRT}{m} = \frac{\kappa nRT}{n \mu} = \frac{\kappa RT}{\mu}
$$
Rychlost zvuku tedy nečekaně závisí na teplotě a molární hmotnosti $\mu$ prostředí, jímž se šíří. 
POZN.: $\kappa$ se oybčejně značí $\gamma$ a představuje stupeň volnosti příslušné molekuly, jíž se zvuk šíří.

## Mody
Máme-li strunu upevněnou ke stěně pro níž platí podmínka, že v místě upevnění, čili v bodě $x=0$ je výchylka nulová, tak v podstatě dáváme naší diferenciální rovnici okrajovou podmínku, čímž omezujeme množinu možných řešení. Aby po doražení ke stěně došlo k "odrazu", tak se musí pohybovat dvě vlny proti vlně od stěny pohybovat nějaká pomyslná vlna, která jí nahradí při vynulování. Výchozí rovnice bude mít tedy tvar:
$$
y = f(x-vt) + G(x+vt)
$$
Položíme-li nyní x rovno nule, tak odsud plyne, že a by byla výchylka rovna nule v počátku, musí být druhá funkce $G(x+vt)$ přesně opačná k funkci $f(x-vt)$, což nám ostatně napovídá i intuice. Tedy:
$$
y = f(x-vt) - f(-x-vt)
$$
Předpokládáme-li sinusoidální charakter vln, můžeme rovnice napsat jako reálnou část z:
$$
f(-x-vt) = e^{i \omega (t+\frac{x}{v})}, f(x-v) = e^{i\omega (t-\frac{x}{v})}
$$
Zkombinujme náš výsledek do prvotní rovnice:
$$
y = e^{i \omega t)}(e^{-\frac{x}{v} \omega i}- e^{-\frac{x}{v} \omega i})
$$
Po úpravě pomocí komplexních čísel:
$$
y = i e^\omega i t sin(\frac{x \omega}{v})
$$
Všimněme si, že všechny body na sinusoidě kmitají se stejnou úhlovou frekvencí a v některých bodech - právě kvůli existenci sinu, bude výchylka nulová. Tyto body se budou opakovat v polohách 

$$
x = \lambda = \frac{2 \pi}{\omega}
$$
Upevníme-li strunu i na druhé straně nějaké pomyslně zdi, tak nám přibyde nová počáteční podmínka naší rovnici. Zohledníme jí právě upravením tvaru funkce uvnitř sinu, protože stejným způsobem jsme zajišťovali anši první počáteční podmínku. Přepíšeme argument sinu do: 
$$
\frac{\omega x}{v} = kx = kL = n\pi
$$
Odsud plyne, že aby toto bylo splěno musí platit (mohli bychom ekvivalentně tvrdit, že funkce je splněná jen pro vybrané rychlosti šíření ale to je podle mě docela zyytečnost):
$$
\omega = \frac{n \pi v}{L}
$$
Mod = periodicky obrazec, který se s časem nějak opakuje 
Vyšší harmonická = frekvence, která je právě celočíselným násobkem základní frekvence.

## Dopplerův jev

Pohybující se zdroj vlnění. Je-li zdroj v klidu platí, že perioda vlnění je $T_{0}$, což $\lambda_{0}=vT_{0}$ 
Zvuk/signál vždycky vypouštíme s určitou frekvencí a když takhle "pliveme" signály při pohybu, tak se ve směru pohybu zkracuje frekvence a proti směru prodlužuje. - všechny se pohybují konstantní rychlostí. Platí tedy obecný vzorec:

$$
f = f_{0} \frac{v+v_{p}}{v+v_{z}} 
$$
Kde v je rychlost šíření vlnění v daném prostředí, $v_{z}$ rychlost zdroje daného vzruchu a $v_{p}$ rychlost pozorovatele. Z obrázku logicky platí, že když se zdroj pohybuje směrem k pozorovateli, tak rychlost přísluší kladné znaménko a naopak záporné.

Speciální případy:

Předpokládáme-li, že se se zachovává množství částic v objemu (mělo by), tak ve starém objemu bude zajisté platit, že:

$$
m = V \rho_0 = \rho_0 \Delta
$$
(předpokládáme jednotkovou plochu kolmou na pohyb vlny, totéž v druhém případě). Poté ale musí notně platit, že bude stejné množství částic i v objemu $V = 1 (x+\Delta)$


# Fourierova řada




# Simulace plné vlnové funkce vs obyčejné.
$$
\frac{\partial^{2}y}{\partial x^{2}} -\frac{1}{c^{2}}\frac{\partial^{2}y}{\partial t^{2}} - \gamma \frac{\partial y}{\partial x} - l^{2} \frac{\partial^{4}y}{\partial x^{4}} = 0
$$
Kde $y$ je funkce šíření vlnění, x je poloha (řešíme v 1D), $t$ je čas a $l$ charakteristická délka. Počáteční podmínky jsou potom standardně $y(0,t)=y(L,t) = 0$, což odpovídá tomu, že máme strunu délky "L" napnutou mezi dvěma bodama. $y(x,0) = f(x)$ 

Tuto rovnici můžeme diskretizovat do nějaké šílené formy - viz zdroj, která potom stav budoucí výchylky počítá rekurzivně z předchozích
$$\frac{y_{j+1}^{m} -2y_j^m + y_{j-1}^{m}}{\Delta x^2} - \frac{1}{c^2}\frac{y_j^{m+1} -2y_j^m + y_j^{m-1}}{\Delta t^2} - \gamma \frac{y_j^{m+1} - y_j^{m-1}}{2 \Delta t} - l^2 \frac{y_{j-2}^m -4y_{j-1}^m +4y_{j}^m -4y_{j+1}^m +y_{j+2}^m}{\Delta x^4} =0 $$
a dále přepsat na:
$$
y_j^{m+1} = \left[\frac{1}{c^2 \Delta t^2} + \frac{\gamma}{2 \Delta t} \right]^{-1} \left[\frac{1}{\Delta x^2} \left( y_{j+1}^{m} -2y_j^m + y_{j-1}^{m} \right) -\frac{1}{c^2 \Delta t^2} \left( y_j^{m-1} - 2y_j^m \right) + \frac{\gamma}{2 \Delta t}y_j^{m-1} -\frac{l^2}{\Delta x^4} \left( y_{j-2}^m -4y_{j-1}^m +6y_{j}^m -4y_{j+1}^m +y_{j+2}^m \right) \right] 
$$

Index m představuje index času na rozdělené časové ose. Jak jsem zmínil na počátku - budoucí index získáváme rekurzivně z předchozího. Diskretizaci používám, abych mohl používat iterativní postup místo nějakého scipy balčíku.

Obyčejná vlnová funkce, tzn. ta, o které jsme se učili má pouze první dva členy, čili:
$$
\frac{\partial^{2}y}{\partial x^{2}} -\frac{1}{c^{2}}\frac{\partial^{2}y}{\partial t^{2}} = 0
$$
Diskretizováno 
$$
y_j^{n+1} = 2y_j^n - y_j^{n-1} + \frac{c^2 \Delta t^2}{\Delta x^2} (y_{j+1}^n - 2y_j^n + y_{j-1}^n)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib.animation import PillowWriter
from numba import jit, njit
from scipy.io import wavfile
from IPython.display import Audio
```

```python
x_sample = 101
t_sample = 500000
L = 0.7 
dx = L/(x_sample-1)
f = 220
c = 2*L*f # nějaký obecný vztah pro rychlost ve struně
dt = 1/t_sample
odtazeni = 5e-6 #nevim, co je tenhle koeficient
gamma = 2.6e-5 #tuhost pružiny
```

```python
# počáteční podmínka pro vytažení
ya = np.linspace(0,0.01,70) 
yb = np.linspace(0.01,0,31)
y0 = np.concatenate([ya,yb]) # spojení dvou listů do jednoho
plt.plot(y0)
```

```python
sol = np.zeros((t_sample,x_sample)) #na každou sekundu připadá informace o poloze všech bodech na pružině
sol[0] = y0
sol[1] = y0 #počáteční podmínky o tom, jak natáhNo nu celou pružinu
```

```python
@numba.jit("f8[:,:](f8[:,:], i8, i8, f8, f8, f8, f8)", nopython=True,nogil=True) 
# specifikuji výstup funkce (flaotová matice), a všechny vstupy funkce
def vypocetnifunkce(d,times,L,dt,dx,l,tuhost):
    for t in range(1,times-1): #začínám od jedničky právě proto, aby mě to nekoplo. Od dvojky protože první dvě máme definované
        for i in range(2, L-2): # iterujeme pro každý čas přes všechny vzdálennosti
            
            prvniclen = ((1/(c**2 * dt**2)) + (tuhost)/(dt*2))**(-1)
            druhyclen = 1/dx**2 * (d[t][i-1] - 2*d[t][i] + d[t][i+1])#(d[t][i+1]- d[t][i] + d[t][i-1]) * (1/dx**2)
            treticlen = 1/(c**2 * dt**2) * (d[t-1][i] - 2*d[t][i])#1/((c**2)*(dt**2))*(d[t-1][i] - 2*d[t][i+1])
            clenctyri = (gamma/(2*dt))*d[t-1][i]
            clenpet = (l**2)/(dx**4) * (d[t][i+2]-4*d[t][i+1]+6*d[t][i-1]+d[t][i-2])
            d[t+1][i] = prvniclen*(druhyclen-treticlen+clenctyri-clenpet)
        
    return d

@numba.jit("f8[:,:](f8[:,:], i8, i8, f8, f8, f8)", nopython=True, nogil=True)
def vypocetni_znama(d,times,L,dt,dx,tuhost):
    
    for t in range(1,times-1):
        
        for i in range(1,L-1):
            d[t+1][i] = 2*d[t][i] - d[t-1][i] + (c**2 * dt**2)/(dx**2)*(d[t][i+1]-2*d[t][i] + d[t][i-1])
    return(d)
```

$$
y_j^{n+1} = 2y_j^n - y_j^{n-1} + \frac{c^2 \Delta t^2}{\Delta x^2} (y_{j+1}^n - 2y_j^n + y_{j-1}^n)
$$

```python
example_solution = vypocetnifunkce(sol,t_sample,x_sample,dt,dx,odtazeni,gamma)
example_solution2 = vypocetni_znama(sol,t_sample,x_sample,dt,dx,gamma)

print(example_solution2)
```

```python
plt.figure(figsize=(20,3))
#plt.plot(example_solution[500])
plt.plot(example_solution[100],label="Presnejsi metoda",color="red")
plt.plot(example_solution2[100],label="Nepresna metoda")
plt.xlabel('Poloha', fontsize=30)
plt.ylabel('Výchylka', fontsize=30)
plt.legend()
```

```python
@njit
def integralharmonic(n, solarray):
    sines = np.sin(n*np.pi*np.linspace(0,1,x_sample)) # x/L reprezentované pomocí linspace, protože ten podíl jde od 0 do jedné žeo.
    return np.array([sum(sines*s)for s in solarray]) # přeintegruji přes všechny řádky, ta část se sumou představuje aproximaci integrálu od 0 do 1.

harmonic_array = [integralharmonic(n,example_solution) for n in range(10)]
harmonic_array2 = [integralharmonic(n,example_solution2) for n in range(10)]
```
