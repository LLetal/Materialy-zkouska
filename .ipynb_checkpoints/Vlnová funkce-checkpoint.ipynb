# Vlnov funkce
Nejprve je dobr si ci, e en vlny v prostoru bude s jistotou popisovat njak funkce tvaru $f(x-ct)$, nebo prv tento tvar zaruuje, e se po urit dob bude opakovat. Stejn funkce, ze kter budeme vychzet je funkce popisujc polohu vlny v zvislosti na ase $\gamma(x,t)$. Souasn za zmnku stoj, e aby mohl zvuk vzniknout, mus bt stedn voln drha mnohem men, ne vzdlennost, na kter se zvuk  - jinak by nemohlo dochzet ke srkm a en vzruchu.
Odvozen vlnov funkce na pkladu zvuku se vychz z t skutenost:
 - I. Plyn se pohybuje a mn se tedy jeho hustota
 - II. Zmn hustoty odpovd zmna tlaku
 - III. Nerovnomrn rozdlen tlaku vyvolv pohyb plynu \
Nco jako dkaz: \

### Druh st

Kdy se plyn pohybuje, tak se mn jeho hustota a s n i tlak. Mme tedy jistou funkci:
$$
p = f(\rho)
$$
Vme, e dodaten tlak, kter zpsobuje pro ns slyiteln vzduch je ve srovnn s rovnovnm tlakem velmi mal. Vztah mezi tlakem a hustotou tedy je:
$$
p_0 + p_e = f(\rho_0 + \rho_e) =  f(\rho_{0}) + \rho_{e}f'(\rho_0)
$$
Nebo kadmu tlaku pslu njak hustota. Protoe je ale zmna t hustoty velmi mal, meme zapsat. Z tohoto vztahu rovn plyne, e nov tlak je pmo mrn vyvolan hustot o njak koeficient $\kappa$, tedy $p_e=\rho_e \kappa$. Tmto jsme odvodili prv druh bod zvislost a meme se pesunout na prvn.
### Prvn st
Je-li posunut vlny v ase t zpsoben zvukovou vlnou $\gamma(x,t)$, tak zan-li se vlna it z bodu x a konec "kapsy" je v  poloze $x +\Delta$, nov poloha potom vlivem en je $x+gamma(x,t)$$x+\Delta x +gamma(x+\Delta,t)$ apod... (viz obrzek.)

Pedpokldme-li, e se se zachovv mnostv stic v objemu (mlo by), tak ve starm objemu bude zajist platit, e:

$$
m = V \rho_0 = \rho_0 \Delta
$$
(pedpokldme jednotkovou plochu kolmou na pohyb vlny, tot v druhm ppad). Pot ale mus notn platit, e bude stejn mnostv stic i v objemu $V = 1 (x+\Delta x + \gamma(x+\Delta x,t) -x - \gamma(x,t))$, pestoe dojde ke zmn hustoty na $\rho$. Bude tedy platit rovnost:
$$
\rho_0 \Delta x = \rho (\Delta x + \gamma(x+\Delta x,t) - \gamma(x,t))
$$
Povimnme si, e len $\gamma(x+\Delta x,t) - \gamma(x,t)$ je definic parciln derivace funkce gamma podle x (pedpokldme, e $\Delta x$ je nesmrn mal). Potom za pedpokladu, e nov hustota je rovna $\rho = \rho_0 + \rho_e$(U v pedchoz sti jsme si to ovili) zskvme:
$$
\rho_0 \Delta x = (\rho_0 + \rho_e)(\frac{\partial \gamma}{\partial x}\Delta x + \Delta x)
$$
Roznsobenm zskme a vykrcenm $\Delta x$ zskme:
$$
\rho_e = -\rho_0 \frac{\partial \gamma}{\partial x} - \rho_e \frac{\partial \gamma}{\partial x}
$$
Znovu vyuijeme skutenosti, e potme s hrozn malmi vcmi, proe druh len vypustme, m zskme:
$$
\rho_e = \rho_0 \frac{\partial \gamma}{\partial x}
$$
Znamnko - je zde proto, e s rostoucm posunutm se hustota peci sniuje, jak se vlna  prostorem.
### III. vztah
Zbv nm najt posledn rovnici. Vezmeme-li opt njak drobn objem vzduchu o dlce $\Delta x$ a ploe o jednotkov velikosti, potom hmotnost tohoto objemu bude:
$$
m = \rho_0 \delta x
$$
Sla psobc na tento mal segment vlny bude mt z 2. newtonova zkona velikost:
$$
F = \rho_0 \Delta x \frac{\partial^2 \gama}{\partial t^2}
$$
V bod x na objem vak psobit sla rovn:
$$
F(x) = p(x,t) 
$$
a proti konci vlny potom njak dal sla:
$$
F(x+ \Delta x) = p(\Delta x +x,t)
$$
(len obsahu vypadv, protoe je jednotkov. ). Po seten sil opt zjiujeme, e jsme nalezli parciln derivace a 2. Newtonv zkon m tvar:
$$
\rho_0 \frac{\partial^2 \gama}{\partial t^2} = \frac{\partial\gamma}{\partial x} p_e
$$

(funkce $p(x,t) $ se pochopiteln skld z dvou st $p_0 $ a $p_e$, nicmn prvn len nm vypadv, nebo je konstantn a parciln derivace jej tedy smae.).
### Zkombinovn vech len
Zkombinujeme-li nyn vechny podkoly dohromady, zskme rovnici:
$$
\frac{\partial^2 \gamma(x,t)}{\partial t^2} \frac{1}{\kappa} = \frac{\partial^2 \gamma(x,t)}{\partial x^2}
$$
Ihned na potku jsme si ukzali, e funkce popisujc pohyb vlny mus notn mt tvar $f(x-vt)$, proe se pokusme do rovnice tuto funkci dosadit, abychom zjistili, zda vyhovuje. CO by to bylo za rovnici funkce, kdyby neodpovdala chovn vlny.
$$
\rho_0 \frac{\partial^2 f(x-vt)}{\partial t^2} = \frac{\partial^2 f(x-vt)}{\partial x^2} 
$$ 

Kdy diferencujeme funkci podle x, tak zskme pouze $f''(x-vt)$, paklie podle t, zskvme vak $v^2 f''(x-vt)$, ili:
$$
f''(x-vt) = \frac{1}{\kappa} v^2 f(x-vt)
$$

Odsud navc plyne, e $\kappa = v^2$, aby rovnost platila. Snadno ovme, e vlnov funkce plat i pro sloen vlmny a stejn tak i pro vlnn pohybujc se opanm smrem.

## Odvozen rychlosti zvuku
UVnit vlny je mezi tlakem a teplotou adiabatick zvislost. Plat tedy:
$$
pV^{\kappa} = konstanta
$$
Paklie je ale hustota ve vztahu k objemu v nepm me, tak pro tlak plat:
$$
p = konstanta \rho ^\kappa
$$
V minul sti jsem si odvodili, e pro rychlost v prosted plat vztah $v = \frac{\partial p}{\partial \rho}$, ili po zderivovn dostvme:
$$
v^2 = \frac{\kappa p}{\rho}
$$
Vyuijeme-li stavov rovnice idelnho plynu $pV = nRT$, tak meme z rovnice dosadit za tlak a dle upravovat, m zskme:
$$
v^2 = \frac{\kappa nRT}{\rho V} = \frac{\kappa nRT}{m} = \frac{\kappa nRT}{n \mu} = \frac{\kappa RT}{\mu}
$$
Rychlost zvuku tedy neekan zvis na teplot a molrn hmotnosti $\mu$ prosted, jm se . 
POZN.: $\kappa$ se oybejn zna $\gamma$ a pedstavuje stupe volnosti pslun molekuly, j se zvuk .

## Mody
Mme-li strunu upevnnou ke stn pro n plat podmnka, e v mst upevnn, ili v bod $x=0$ je vchylka nulov, tak v podstat dvme na diferenciln rovnici okrajovou podmnku, m omezujeme mnoinu monch een. Aby po doraen ke stn dolo k "odrazu", tak se mus pohybovat dv vlny proti vln od stny pohybovat njak pomysln vlna, kter j nahrad pi vynulovn. Vchoz rovnice bude mt tedy tvar:
$$
y = f(x-vt) + G(x+vt)
$$
Polome-li nyn x rovno nule, tak odsud plyne, e a by byla vchylka rovna nule v potku, mus bt druh funkce $G(x+vt)$ pesn opan k funkci $f(x-vt)$, co nm ostatn napovd i intuice. Tedy:
$$
y = f(x-vt) - f(-x-vt)
$$
Pedpokldme-li sinusoidln charakter vln, meme rovnice napsat jako relnou st z:
$$
f(-x-vt) = e^{i \omega (t+\frac{x}{v})}, f(x-v) = e^{i\omega (t-\frac{x}{v})}
$$
Zkombinujme n vsledek do prvotn rovnice:
$$
y = e^{i \omega t)}(e^{-\frac{x}{v} \omega i}- e^{-\frac{x}{v} \omega i})
$$
Po prav pomoc komplexnch sel:
$$
y = i e^\omega i t sin(\frac{x \omega}{v})
$$
Vimnme si, e vechny body na sinusoid kmitaj se stejnou hlovou frekvenc a v nkterch bodech - prv kvli existenci sinu, bude vchylka nulov. Tyto body se budou opakovat v polohch 

$$
x = \lambda = \frac{2 \pi}{\omega}
$$
Upevnme-li strunu i na druh stran njak pomysln zdi, tak nm pibyde nov poten podmnka na rovnici. Zohlednme j prv upravenm tvaru funkce uvnit sinu, protoe stejnm zpsobem jsme zajiovali ani prvn poten podmnku. Pepeme argument sinu do: 
$$
\frac{\omega x}{v} = kx = kL = n\pi
$$
Odsud plyne, e aby toto bylo splno mus platit (mohli bychom ekvivalentn tvrdit, e funkce je splnn jen pro vybran rychlosti en ale to je podle m docela zyytenost):
$$
\omega = \frac{n \pi v}{L}
$$
Mod = periodicky obrazec, kter se s asem njak opakuje 
Vy harmonick = frekvence, kter je prv celoselnm nsobkem zkladn frekvence.

# Fourierova ada




# Simulace pln vlnov funkce vs obyejn.
$$
\frac{\partial^{2}y}{\partial x^{2}} -\frac{1}{c^{2}}\frac{\partial^{2}y}{\partial t^{2}} - \gamma \frac{\partial y}{\partial x} - l^{2} \frac{\partial^{4}y}{\partial x^{4}} = 0
$$
Kde $y$ je funkce en vlnn, x je poloha (eme v 1D), $t$ je as a $l$ charakteristick dlka. Poten podmnky jsou potom standardn $y(0,t)=y(L,t) = 0$, co odpovd tomu, e mme strunu dlky "L" napnutou mezi dvma bodama. $y(x,0) = f(x)$ 

Tuto rovnici meme diskretizovat do njak len formy - viz zdroj, kter potom stav budouc vchylky pot rekurzivn z pedchozch
$$\frac{y_{j+1}^{m} -2y_j^m + y_{j-1}^{m}}{\Delta x^2} - \frac{1}{c^2}\frac{y_j^{m+1} -2y_j^m + y_j^{m-1}}{\Delta t^2} - \gamma \frac{y_j^{m+1} - y_j^{m-1}}{2 \Delta t} - l^2 \frac{y_{j-2}^m -4y_{j-1}^m +4y_{j}^m -4y_{j+1}^m +y_{j+2}^m}{\Delta x^4} =0 $$
a dle pepsat na:
$$
y_j^{m+1} = \left[\frac{1}{c^2 \Delta t^2} + \frac{\gamma}{2 \Delta t} \right]^{-1} \left[\frac{1}{\Delta x^2} \left( y_{j+1}^{m} -2y_j^m + y_{j-1}^{m} \right) -\frac{1}{c^2 \Delta t^2} \left( y_j^{m-1} - 2y_j^m \right) + \frac{\gamma}{2 \Delta t}y_j^{m-1} -\frac{l^2}{\Delta x^4} \left( y_{j-2}^m -4y_{j-1}^m +6y_{j}^m -4y_{j+1}^m +y_{j+2}^m \right) \right] 
$$

Index m pedstavuje index asu na rozdlen asov ose. Jak jsem zmnil na potku - budouc index zskvme rekurzivn z pedchozho. Diskretizaci pouvm, abych mohl pouvat iterativn postup msto njakho scipy balku.

Obyejn vlnov funkce, tzn. ta, o kter jsme se uili m pouze prvn dva leny, ili:
$$
\frac{\partial^{2}y}{\partial x^{2}} -\frac{1}{c^{2}}\frac{\partial^{2}y}{\partial t^{2}} = 0
$$
Diskretizovno 
$$
y_j^{n+1} = 2y_j^n - y_j^{n-1} + \frac{c^2 \Delta t^2}{\Delta x^2} (y_{j+1}^n - 2y_j^n + y_{j-1}^n)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib.animation import PillowWriter
from numba import jit, njit
from scipy.io import wavfile
from IPython.display import Audio
```

```python
x_sample = 101
t_sample = 500000
L = 0.7 
dx = L/(x_sample-1)
f = 220
c = 2*L*f # njak obecn vztah pro rychlost ve strun
dt = 1/t_sample
odtazeni = 5e-6 #nevim, co je tenhle koeficient
gamma = 2.6e-5 #tuhost pruiny
```

```python
# poten podmnka pro vytaen
ya = np.linspace(0,0.01,70) 
yb = np.linspace(0.01,0,31)
y0 = np.concatenate([ya,yb]) # spojen dvou list do jednoho
plt.plot(y0)
```

```python
sol = np.zeros((t_sample,x_sample)) #na kadou sekundu pipad informace o poloze vech bodech na pruin
sol[0] = y0
sol[1] = y0 #poten podmnky o tom, jak nathNo nu celou pruinu
```

```python
@numba.jit("f8[:,:](f8[:,:], i8, i8, f8, f8, f8, f8)", nopython=True,nogil=True) 
# specifikuji vstup funkce (flaotov matice), a vechny vstupy funkce
def vypocetnifunkce(d,times,L,dt,dx,l,tuhost):
    for t in range(1,times-1): #zanm od jedniky prv proto, aby m to nekoplo. Od dvojky protoe prvn dv mme definovan
        for i in range(2, L-2): # iterujeme pro kad as pes vechny vzdlennosti
            
            prvniclen = ((1/(c**2 * dt**2)) + (tuhost)/(dt*2))**(-1)
            druhyclen = 1/dx**2 * (d[t][i-1] - 2*d[t][i] + d[t][i+1])#(d[t][i+1]- d[t][i] + d[t][i-1]) * (1/dx**2)
            treticlen = 1/(c**2 * dt**2) * (d[t-1][i] - 2*d[t][i])#1/((c**2)*(dt**2))*(d[t-1][i] - 2*d[t][i+1])
            clenctyri = (gamma/(2*dt))*d[t-1][i]
            clenpet = (l**2)/(dx**4) * (d[t][i+2]-4*d[t][i+1]+6*d[t][i-1]+d[t][i-2])
            d[t+1][i] = prvniclen*(druhyclen-treticlen+clenctyri-clenpet)
        
    return d

@numba.jit("f8[:,:](f8[:,:], i8, i8, f8, f8, f8)", nopython=True, nogil=True)
def vypocetni_znama(d,times,L,dt,dx,tuhost):
    
    for t in range(1,times-1):
        
        for i in range(1,L-1):
            d[t+1][i] = 2*d[t][i] - d[t-1][i] + (c**2 * dt**2)/(dx**2)*(d[t][i+1]-2*d[t][i] + d[t][i-1])
    return(d)
```

$$
y_j^{n+1} = 2y_j^n - y_j^{n-1} + \frac{c^2 \Delta t^2}{\Delta x^2} (y_{j+1}^n - 2y_j^n + y_{j-1}^n)
$$

```python
example_solution = vypocetnifunkce(sol,t_sample,x_sample,dt,dx,odtazeni,gamma)
example_solution2 = vypocetni_znama(sol,t_sample,x_sample,dt,dx,gamma)

print(example_solution2)
```

```python
@njit
def integralharmonic(n, solarray):
    sines = np.sin(n*np.pi*np.linspace(0,1,x_sample)) # x/L reprezentovan pomoc linspace, protoe ten podl jde od 0 do jedn eo.
    return np.array([sum(sines*s)for s in solarray]) # peintegruji pes vechny dky, ta st se sumou pedstavuje aproximaci integrlu od 0 do 1.

harmonic_array = [integralharmonic(n,example_solution) for n in range(10)]
harmonic_array2 = [integralharmonic(n,example_solution2) for n in range(10)]
```
