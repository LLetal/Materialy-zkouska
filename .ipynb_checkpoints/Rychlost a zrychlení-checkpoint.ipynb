# Rychlost a zrychlení

Okamžitá rychlost je vektorová veličina, kterou spočítáme jako:
$$
\vec{v}= \lim_{x \to x_0} \frac{\Delta \vec{r}}{\Delta t}= \frac{d\vec{r}}{dt} = \dot r 
$$
Od ní odvozené zrychlení $\vec{a}$ poté je změnou v čase, čili:
$$
\vec{a} = \frac{d\vec{v}}{dt} = \ddot r 
$$

# Normálnové a tečné zrychlení
Zrychlení můžeme - nejen u kruhového pohybu - rozdělit na dvě složky. ** Tečnou a normálovou**
Tečné zrychlení vypočítáme pomocí vztahu:
$$
\vec{a_t} = \frac{\vec{a}\vec{v}}{|v|} \frac{\vec{v}}{|v|}, a_t = \frac{dv}{dt}
$$
Má stejný směr jako rychlost otáčejícího se tělesa  
Naproti tomu normálové spočítáme jako:
$$
\vec{a_n} = \frac{\vec{a}\vec{R}}{R} \frac{\vec{R}}{R}, a_n = \frac{v^2}{R}
$$
A má směr kolmý k pohybu tělesa.
