## hydrostatika a hydromechanika
Pro tekutinu plat, e:
$$
\sigma_{ii} = -p, p>0, \tau_{ij} = 0
$$
esky: Ideln kapalina = nestlaiteln, pokud je kapalina v klidu jsou smykov napt nulov, co d rozum.
### Pascalv zkon
Tlak vyvolan vnj silou, kter psob na kapalinu v uzaven ndob je ve vech mstech kapaliny stejn. Tto skutenosti vyuvme napklad u hydraulickch lis, potako zvedacch kesel. 

### Odvozen hydrostatickho tlaku
Uvaujme nekonen mal dleek uvnit kapaliny, mli by na nj psobit celkem 3 mon sly. Jednou z nich je thov sla, kter psob na cel objem buky. Na kostiku mimo to psob jet dv plon sly - shora a zdola. Shora psob na kostiku sla pipadajc ostatn vod nad nm a zespodu reakn sla kostiky pod nm. Protoe uvaujeme, e kostika je v klidu, mus notn platit:
$$
F_1 + F_g + F_2 = 0
$$

Schmaticky. Osu si zvolme stejn jako na obrzku. Potom tedy:
$$
F_1 = -p(z+dz)dS
$$
$$
F_2 = p(z)dS
$$
$$
F_g = - dm g = - dV \rho g
$$
Dosame do pvodn rovnice:
$$
-p(z+dz)dS +p(z)dS  - dS dz \rho g = 0
$$

$$
\frac{p(z)-p(z+dz)}{dz} - \rho g = 0
$$

$$
\frac{\partial p}{\partial z} = - \rho(z) g
$$
Nyn pokud chceme zjistit rozdl tlaku mezi njakmi dvma body, tak jednodue provedeme integrl z dvou pslunch poloh - referenn teba hladinu a njak bod dole. ili:
$$
\Delta p = -\rho g [g]_{z}^{z_h} = - \rho g(z_h-z) = - \rho g(0-z) = \rho g h
$$

Pozn.: Hustota nemus bt konstantn, u ideln kapaliny ale konstantn je. Bon leny nezapotvme do rovnosti proto, e se vynuluj - jsou ve stejn hloubce, nicmn maj vdy opan smry. Funkce tlaku, tak jak j mme definovanou, zvis na hloubce a proto bude sla jin zespodu tlesa jako nahoe a nevynuluj se.


### Archimdv zkon
Tleso ponoen v kapalin je nadlehovno silou, kter se rovn tze tekutiny o stejnm objemu jako ponoen st tlesa.
M-li platit, e kapalina je v klidu, mus notn platit rovnovha sil:

$$
F_2 - F_1 - F_g = 0
$$
V pedchozm buce jsme si vak odvodili, e rozdl sil $F_2$ a $F_1$ je roven: 
$$
\rho_V h_V S g
$$
a tha tlesa:
$$
F_g = \rho h S g
$$
Dosazenm zskvme:
$$
\rho_V h_V = \rho h
$$
Jinak by to pece neplavalo a u by soustava nebyla v klidu.
