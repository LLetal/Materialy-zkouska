## hydrostatika a hydromechanika
Pro tekutinu platí, že:
$$
\sigma_{ii} = -p, p>0, \tau_{ij} = 0
$$
Česky: Ideální kapalina = nestlačitelná, pokud je kapalina v klidu jsou smyková napětí nulová, což dá rozum.
### Pascalův zákon
Tlak vyvolaný vnější silou, která působí na kapalinu v uzavřené nádobě je ve všech místech kapaliny stejný. Této skutečnosti využíváme například u hydraulických lisů, potažko zvedacích křesel. 

### Odvození hydrostatického tlaku
Uvažujme nekonečně malý díleček uvnitř kapaliny, měli by na něj působit celkem 3 možné síly. Jednou z nich je tíhová síla, která působí na celý objem buňky. Na kostičku mimo to působí ještě dvě plošné síly - shora a zdola. Shora působí na kostičku síla připadající ostatní vodě nad ním a zespodu reakční síla kostičky pod ním. Protože uvažujeme, že kostička je v klidu, musí notně platit:
$$
F_1 + F_g + F_2 = 0
$$

Schématicky. Osu si zvolme stejně jako na obrázku. Potom tedy:
$$
F_1 = -p(z+dz)dS
$$
$$
F_2 = p(z)dS
$$
$$
F_g = - dm g = - dV \rho g
$$
Dosaďme do původní rovnice:
$$
-p(z+dz)dS +p(z)dS  - dS dz \rho g = 0
$$

$$
\frac{p(z)-p(z+dz)}{dz} - \rho g = 0
$$

$$
\frac{\partial p}{\partial z} = - \rho(z) g
$$
Nyní pokud chceme zjistit rozdíl tlaku mezi nějakými dvěma body, tak jednoduše provedeme integrál z dvou příslušných poloh - referenčně třeba hladinu a nějaký bod dole. Čili:
$$
\Delta p = -\rho g [g]_{z}^{z_h} = - \rho g(z_h-z) = - \rho g(0-z) = \rho g h
$$

Pozn.: Hustota nemusí být konstantní, u ideální kapaliny ale konstantní je. Boční členy nezapočítáváme do rovnosti proto, že se vynulují - jsou ve stejné hloubce, nicméně mají vždy opačné směry. Funkce tlaku, tak jak jí máme definovanou, závisí na hloubce a proto bude síla jiná zespodu tělesa jako nahoře a nevynulují se.


### Archimédův zákon
Těleso ponořené v kapalině je nadlehčováno silou, která se rovná tíze tekutiny o stejném objemu jako ponořená část tělesa.
Má-li platit, že kapalina je v klidu, musí notně platit rovnováha sil:

$$
F_2 - F_1 - F_g = 0
$$
V předchozím buňce jsme si však odvodili, že rozdíl sil $F_2$ a $F_1$ je roven: 
$$
\rho_V h_V S g
$$
a tíha tělesa:
$$
F_g = \rho h S g
$$
Dosazením získáváme:
$$
\rho_V h_V = \rho h
$$
Jinak by to přece neplavalo a už by soustava nebyla v klidu.
