# Steinerova věta
Máme-li dvě rovnoběžné osy, tak momenty setrvačnosti vůči těmto dvou osám jsou ve vztahu:

$$
J = J_{T} + MR_{T}^{2}
$$

## Moment setrvačnosti u kolmých os
zjistit moment setrvačnosti vzhledem k nějaké ose, na kterou jsou kolmé dvě další, platí vztah:
$$
J_{z} = J_{x} + J_{y}
$$
## Chaslesův teorém
Libovolný pohyb tuhého tělesa lze složit z rotace kolem osy a translace podél této osy
$$
v = v_{R} + v_{T}
$$

## Moment setrvačnosti vzhledem k obecné ose
Využíváme tzv. tenzor setrvačnosti.

$$
\begin{pmatrix}
J_{xx} & D_{xy} & D_{xz}\\
D_{xy} & J_{yy} & D_{yz}\\
D_{xz} & D_{yz} & J_{zz}
\end{pmatrix}
$$


První člen na diagonále představuje vzdálennost od osy x, druhý od osy y, třetí od osy z. Zbylé členy od příslušných ploch(např $D_{xy}$ je vzdálenost od roviny xy).

### Další poznámky k setrvačnosti
- Skládá-li se těleso z více těles, jejichž momenty setrvačnosti známe, tak celkový moment setrvačnosti získáme jako jejich součet
- Moment setrvačnosti vzhledem ke kterékoliv ose je roven součtu momentu setrvačnosti vzhledem k ose procházející hmotným středem a součinu celkové hmotnosti a druhé mocniny vzdálenosti od osy hmotného středu (Steiner)
- Má-li těleso tvar rovinného obrazce je moment setrvačnosti vzhledem k ose kolmé na jeho rovinu roven součtu setrvačnosti kterých-koliv dvou kolmých os ležicích v rovině a protínajících se na kolmé ose.
## Setrvačník



Když při translačním pohybu na těleso s nějakou hybností působíme silou F, tak se za touto silou začne pohybovat. U setrvačníku použijeme příslušné veličiny (Moment síly a moment hybnosti), v takovém případě též bude těleso otáčet do směru momentu síly.

## Precese (změna osy otáčení)
Bez působení momentu síly a s působením momentu síly. V první případě je změna osy otáčení způsobená nestálým momentem setrvačnosti (-> v důsledku nedokonalosti tělesa). Rychlost, s jakou se osa mění v tomto případě spočítáme jako:
$$
\omega_p = \frac{I_s \omega_s}{I_p cos(\aplha)}
$$
Kde $I_s$ je moment setrvačnosti vzhledem k ose otáčení, $\omega_s$ úhlová rychlost otáčení vzhledem k této ose, $I_p$ moment setrvačnosti k přidruženým osám a $\alpha$ úhel momentu setrvačnosti vzhleedem k ose symetrie.

```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
from scipy.integrate import dblquad
from scipy.integrate import tplquad
import pandas as pd
```

```python
def f(angle): # funkce poloměru točící káči v závislosti na úhlu
    return(1-np.pi/4 +angle**(1/10))
```

```python
angles=  np.linspace(np.pi/4,np.pi/2,1000)
r = f(angles)
x_souradnice = np.insert(r*np.cos(angles),0,0)
y_souradnice = np.insert(r*np.sin(angles),0,0)

plt.plot(x_souradnice,y_souradnice)
plt.grid()
plt.ylim(-1.5,1.5)
plt.xlim(-1.5,1.5)
```
