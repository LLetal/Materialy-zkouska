# Newtonovy zákony a vrhy

## Hybnost
- míra posubného pohybu tělesa
$$
\vec{p} = m \vec{v}
$$
## Zákon setrvačnosti
Těleso zůstává v klidu nebo v rovnoměrném přímočarém pohybu, pakliže není nuceno vnějšími silami tento svůj stav změnit.
$$
\vec{F} = 0 <=> \frac{d \vec{v}}{dt} = 0
$$
## Zákon síly
Síla odpovídá časové změně hybnosti:
$$
\vec{F} = \frac{d \vec{p}}{dt} = \frac{d}{dt} m \vec{v}
$$
## Zákon akce a reakce
$$
\vec{F_{A,B}} = \vec{F_{B,A}} 
$$

# Odpor vzduchu
Je dán vztahem:

$$
F_{o} = -\frac{1}{2} \rho S C_{d}(Re) v^{2} \frac{\vec{v}}{v} 
$$
Kde S představuje průřez tělesa, $\rho$ hustotu prostředí $v$ relativní rychlost,$C_{d}$ koeficient odporu. Koeficient závisí na Reynoldsově čísle, které je úměrné:

$$
Re = \frac{v L \rho}{\mu}
$$
Kde $L$ je charakteristická rozměr tělesa - něco jako průřez do směru pohybu, $\rho$ hustota prostředí, $\mu$ viskozita - míra toho, jak moc se prostředí brání pohybu.


Při laminárním proudění ($Re<10^{3}$) -> $C_d = \frac{1}{v}$) - Stoeks
Při turbulentním proudění
($10^{5}<Re<10^{5}$) -> $C_d = konst$ - Newton
Při odporu vzduchu pozorujeme  terminální ryclost, což je maximální rychlost, které při pádu dosáhneme - v důsledku odporu vzduchu.
Charakteristický rozměr v kontextu(=rozměr který stojí proti obtékání) Reynoldsova čísla na příkladech:\
U koule = průměr
\
U tyče = průměr tyče
\
U člověka = jeho výška 
\
U křídla = jeho délka 


# Impuls síly
Udává, míru změny síly v závislosti na čase:
$$
I = F \Delta t
$$
Pokud síla není konstantní, pak platí:
$$
I = \int_{t_{1}}^{t_{2}} F dt
$$
Nejčastěji při nárazu/velmi ve velmi krátkých časových intervalech. Jako $\Delta t $ se často doplňuje rychlost ššíření vzhruchu materiálem. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.optimize import minimize
```

```python
def volna_draha(x0,v0, tmax,t_sample=100):
    g = np.array([0,-9.81])
    seznam_casu = np.linspace(0, tmax, t_sample)
    xy = [ x0 + v0 * t + 0.5 * g * t**2 for t in seznam_casu ]
    xy = np.array(xy)
    return(xy)
def volna_draha(t,U):
    g =-9.81
    x,y,vx,vy = U
    
    ax = 0
    ay = g
    return(np.array([vx,vy,ax,ay]))

def zlovolna_draha(t,U):
    g =-9.81
    x,y,vx,vy = U
    Cd = 0.5
    a = 2e-2
    rho = 1.26
    v = np.sqrt(vy**2 + vx**2)
    m = 2.7e-3
    ax = -(1/2) * Cd * np.pi * (a**2) * rho * v*vx/m
    
    ay = g -(1/2) * Cd * np.pi * (a**2) * rho * v*vy/m   
    
    return(np.array([vx,vy,ax,ay]))

def zerochecker(t,U):
    x,y,vx,vy = U
    return(y+0.001)

zerochecker.terminal = True
v0 = 50
def optimalizace(angle):
    tmax = 10
    vx = np.cos(angle)*v0

    vy = np.sin(angle)*v0
    
    x = 0
    y = 0
    V0 = np.array([x,y,vx[0],vy[0]])
    x_result = -solve_ivp(zlovolna_draha,(0,tmax),V0,max_step = 1/100,events=zerochecker).y[0][-1]
    return(x_result) 
```

```python
# pocatecni podminky

v0 = 50

phis = [np.radians(angle) for angle in range(10,90,10)]

vx = [np.cos(angle)*v0 for angle in phis]

vy = [np.sin(angle)*v0 for angle in phis]

x = np.zeros(len(vx))
y = np.zeros(len(vx))
solutions1 = []
solutions2 = []
tmax = 10
for index in range(0,len(phis)):
    V0 = np.array([x[0],y[0],vx[index],vy[index]])
    solutions2.append(solve_ivp(zlovolna_draha,(0,tmax),V0,max_step = 1/100,events=zerochecker))
    solutions1.append(solve_ivp(volna_draha,(0,tmax),V0,max_step = 1/100,events=zerochecker))
```

```python
xaxis = [x for x in range(int(solutions1[-1].y[0][-1]))] 
for i in range(len(solutions1)-7):
    plt.plot(solutions1[i].y[0],solutions1[i].y[1],label = f"Volná při úhlu {np.rad2deg(phis[i]):.0f}°")
    plt.plot(solutions2[i].y[0],solutions2[i].y[1],label = f"Zlovolná při úhlu {np.rad2deg(phis[i]):.0f}°")
    plt.plot()
    
plt.rcParams['figure.figsize'] = [12, 8]
# plt.plot(xaxis,np.zeros(int(solutions1[-1].y[0][-1])),c="black")
plt.legend()
```
