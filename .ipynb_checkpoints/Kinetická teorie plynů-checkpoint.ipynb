# Kinetická teorie plynů
Bohužel nedokážeme popsat vlastnosti látek, a tedy i plynů, nějakou teorií podobnou např. Newtonovým zákonům a místo toho se spoléháme na statistiku a velkou míru aproximace. Jedinný způsob, jak dokážeme rozumně aproximovat chování např. kapalin/plynů je pomocí numerických simulací, které simulují chování každé částice samostatně. My se naštěstí tomuto věnovat nemusíme a spokojíme se pouze s aproximacemi. Podotýkám ještě, že teorie plynů se často pohybuje na pomezí mikrosvěta nebo tak něco.

# Tlak plynu
Vezměme nějakou pomyslnou nádobku naplněnou plynem na jejíž straně se nachází pohyblivý píst, za nímž je vakuum. Uvnitř této nádobky se pohybuje nespočet molekul s rozličnými rychlostmi, které pístu při každé srážce dodají nějakou nepatrnou hybnost, v důsledku čehož se píst čase začne posunovat. Tlak plynu láhve potom odpovídá síle, kteoru musíme na láhev onen píst o obsahu $S$ působit, abychom zabránili posunování.
$$
p = \frac{F}{S}
$$
Právě když posuneme pístem o určitou vzdálenost $dx$ vykonáme práci:
$$
dW = Fdx = p dV
$$
Celkovou práci vykonnanou plynem potom snadno integrálem. V případě, že by měli všechny molekuly pouze x-ovou složku rychlosti, můžeme ekvivalentně odvodit vztah pro velikost síly, kterou působí na píst. Nejprve uvažme, že př ikaždé srážce se zdí příslušná molekula od zdi odletí se stejnou hybností jen opačného směru, pročež předá stěně hybnost $2mv_x$ (ze zzH 2p - p = p0). Uvnitř nádoby je ale těchto molekul nějaký počet a z něj mohou za jednu sekundu do pístu narazit právě ty částice, které jsou pouze ve vzdálenosti $v_x 1$. V této vzdálennosti se nachází celkem N částic, jejichž počet je dán vztahem:
$$
N = n_V S v_x
$$
Celková hybnost, kterou částice plynu předají pístu(síla) tedy je:
$$
F = 2n_V m S v_x^2
$$

## Adiabatické stlačení plynu
Všechna práce, kterou dodáme pístu tím, že ho stlačíme se využije na urychlení molekul plynu uvnitř válce.
$$
pV = (\gamma -1)U
$$
Kde $\gamma$ představuje stupně volnosti příslušného plynu uvnitř nádoby a $U$ jeho vnitřní energii. Tudíž platí:
$$
pdV = -dU
$$
(Všechno se přeměná na zvýšení energie). Z vrchního vztahu platí, že $U = \frac{pV}{\gamma -1}$, zderivováním a dosazením do vrchní rovnice:
$$
pdV = \frac{-(pdV dpV)}{\gamma-1}
$$
Po upravení:
$$
\frac{dV}{V} (\gamma) + \frac{dp}{p} = 0 = konst
$$
Rovnici dále integrujeme a získáme nejprve nějaký výraz s logaritmy, po převodu na exponenciály však:
$$
pV^\gamma = konst
$$
