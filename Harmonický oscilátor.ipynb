## Harmonický oscilátor

Nejpoužívanější koncept ve fyzice. Aplikace po celé fyzice od mechaniku po kvantovou fyziku. Intuitivně vycházíme ze závaží na  pružině.  

$$
\frac{d^{2}x}{dt^{2}} + \frac{k}{m}x = 0
$$

$$
\frac{d^{2}x}{dt^{2}} + \omega_{0}^{2}x = 0
$$
Jedná se o jednoduchou diferenciální rovnici, jejíž řešení je od pohledu nějaký sinus, můžeme jí však řešit i standardním postupem přes $e^{\lambda x}$.
$$x(t) = C_{1} cos(\omega_{0}t+\phi_{1})+C_{2} sin(\omega_{0}t + \phi_{2})
$$
Výsledek můžeme v závislosti na počátečních podmínkách zapsat třeba jako:
$$
x(t) = A sin(\omega_{0} t + \alpha)
$$
Důležitá  identita je:
$$
cos(\Theta) = sin(\Theta +\frac{\pi}{2})
$$
Řešení konkrétního případu může tedy vypadat např. takto (při $x(t=0)= -A, v_{x}(t=0)$):

$$
x(t) = -A cos(\sqrt{\frac{k}{m}}t + \phi)
$$
Perioda kmitů tohoto oscilátoru potom je:

$$
T = 2\pi \sqrt{\frac{m}{k}}
$$

# Práce vykonaná harmonickým oscilátorem
Práce při pohybu z bodu A do bodu B
$$
W_{AB} = \int_{x_{A}}^{x_{B}} -kx dx = -\frac{1}{2}(x_{B}^{2}-x_{A}^{2})
$$
Vztáhneme-li práci k přesunutí z počátku (rovnovážného stavu, kde vychýlení = 0) do nějakého bodu, získáme:
$$
W_{PA} = - \frac{1}{2} k x_{A}^{2} 
$$
Což představuje potenciální energii. Kinetická energie tohoto oscilátoru potom je:

$$
E_{k} = \frac{1}{2} m \frac{dx}{dt} = \frac{1}{2} m \omega^{2} cos (\omega t +\phi)
$$


# Fyzické kyvadlo - speciální případ harmonického oscilátoru

Fyzické kyvadlo je těleso jakého-koliv tvaru rotující kolem nějaké osy. Pro pohyb tohot kyvadla potom s jistotou platí 2. impulsová věta, a tedy:
$$
\tau_{0} = Mgl sin(\phi) = -J_{0} \frac{d^{2}\phi}{dt^{2}}
$$
Při apromixami malých kmitů ($\phi <<5°$) potom získáváme rovnici harmonického oscilátoru. Často zde užíváme tzv. Steinerovku, protože bude osa otáčení čas mimo těžiště tělesa, ale vždy alespoň rovnoběžná s tou procházející hmotným středem. Slide z prezentace


# Tlumené kmity
Trochu realističtější harmonický oscilátor. Přibývá v něm člen brzdící pohyb, který je přímo úměrný rychlosti - docela realistický předpoklad. Pro pružinu tedy platí, že:

$$
m \ddot x =  -kx -h \dot x
$$
Po úpravě:
$$
m \ddot x + kx +h \dot x = 0, \ddot x +   2 \delta \dot x +\omega^{2} x
$$

$$
\omega = \sqrt{\frac{k}{m}}
$$
## Buzené kmity

Udržuje-li oscilátor v pohybu nějaká budící síla platí, že:
$$
\ddot x + \omega_{0}^{2} x = F_{0} cos(\omega t + \Delta)
$$
A v případě tlumených kmitů:
$$
\ddot x + 2 \delta \dot x+ \omega_{0}^{2} x = F_{0} cos(\omega t + \Delta)
$$
Při řešení rovnic tohoto typu využíváme Eulerovy identity $e^{i\omega t+\Delta}=cos(\omega +\delta) + isin(\omega +\delta)$, přičemž z ní vždy ve výsledku bereme jen reálnou část, protože fyzikální jevy nemohou mít imaginární podstatu. Nyní si demonstrujme sílu tohoto postupu na příkladu netlumených buzených kmitů. Pravou stranu můžeme přepsat do tvaru:
$$
\ddot x + \omega_{0}^{2} x = F_{0} cos(\omega t + \Delta) = F_{0}e^{i\omega t + \Delta} = \hat F e^{i \omega t}
$$
Kde $\hat F$ představuje tzv. komplexní amplitudu. Nyní víme, že výsledný oscilátor bude obecně kmitat se stejnou úhlovou frekvencí jako budící síla (až po nějakém čase, kdy frekvence budící síly překoná vlastní frekvenci, ale logicky dává smysl, že bude kmitat se stejnou úhlovou frekvencí) ale obecně jinou fází. Z obecné rovnice řešení harmonického oscilátoru víme, že jeho řešení je reálnou částí z $e^{i \omega t + \Phi}$, čili $x = A cos(\omega t + \Phi)=  A sin(\omega t + \Phi_{2})$ (mínus je v řešení z důvodu identity $cos(x -\frac{\pi}{2}) = sin(x)$ Nyní, když víme, že bude mít rovnice přibližně tvar tohoto řešení a máme stanovené předpoklady, vidíme, že máme v podstatě rovnici o jedné neznáme ve tvaru:
$$
\hat x \frac{d^{2}e^{i \omega t}}{dt^{2}} + \omega_{0}^{2} \hat x e^{i \omega t} = \hat F e^{i \omega t}
$$
Odsud můžeme po zderivování vykrátit onen člen $e^{i \omega t }$ (při derivování se nijak nemění) a získáme konečnou rovnici pro $\hat x$:
$$
\hat x (i \omega)^{2} + \omega_{0}^{2}\hat x = \hat F
$$
$$
\hat x = \frac{\hat F}{\omega_{0}^{2} - \omega^{2}} = F_{0} e^{-\Delta} \frac{1}{\omega_{0}^{2}- \omega^{2}}
$$
$$
x_{p} = Re [\hat x e^{i \omega t}] = F_{0} cos(\omega t + \Delta + \theta) \frac{1}{\omega_{0}^{2} - \omega^{2}}
$$
Protože je zlomek již v reálné části, netřeba se jej při převodu do reálné zbavovat násobením komplexně sdruženým číslem, což u příkladu tlumených buzených kmitů musíme udělat. POZN.: $\theta$ představuje fázový posun oscilátoru, kdyby nebyl buzený - alespoň si to myslím.
Obecné řešení potom má tvar:
$$
x = x_{H} + x_{p} = A cos(\omega t + \phi) + F_{0} cos(\omega t + \Delta + \phi) \frac{1}{\omega_{0}^{2} - \omega^{2}} 
$$

## Buzené kmit pro případ tlumených kmitů
Vycházíme z rovnice:
$$
\ddot x + 2 \delta \dot x+ \omega_{0}^{2} x = F_{0} cos(\omega t + \Delta)
$$
Pro tu můžeme snadno najít homogenní řešení. Očekávané partikulární řešení bude mít tedy stejný tvar, jako to u předchozího oscilátoru. Získáme tedy opět rovnost:
$$
\hat x (i \omega)^{2} + \hat x \omega i + \omega_{0}^{2}\hat x = \frac{\hat F}{m}
$$
Tedy dále znovu: 
$$
\hat x = \frac{\hat F}{m (\omega_{0}^{2} - \omega^{2} + i 2\delta \omega)}
$$
Obecné řešení bude mít tedy opět tvar reálné části z tohoto výrazu zkombinovaného s $e^{i \omega t}$,čili:
$$
x_{p} = F_{0} cos(\omega t + \Delta + \Phi) \frac{1}{m^{2} (\omega_{0}^{2} - \omega^{2} + i 2\delta \omega)^{2}} = F_{0} cos(\omega t + \Delta + \Phi) \frac{1}{m \sqrt{(\omega_{0}^{2} - \omega^{2})^{2} - 4\delta^{2} \omega^{2}}}, \Phi = tan(\frac{2\delta \omega}{\omega_{0}^{2}-\omega^{2}})
$$
Znaméno $\Phi$ by mělo být vždy záporné, protože vždy osciáltor bude vždy zaostávat za budící silou - nějaká setrvačnost. Pod odmocninou to je proto, že velikost získám jako ODMOCNINU, druhých mocnin reálné a imaginární časti!. Rezonanční frekvenci, tzn. frekvenci, při které budou největš výchylky získám derivací amplitudy podle její proměnné, tzn. $\omega$. V případě, že vstoupíme do tlumených kmitů, tak se k amplitudě $A$ ještě přidává člen $e^{-\delta t }$, který vyvstává z důvodu komplexního řešení a představuje efekt TLUMENÍ.

## Energie oscilátoru

Důvod, proč energie s čase klesá je ten, že se jedná o TLUMENÉ kmity, proti jejichž pohybu působí právě koeficient$h$, pročež energie klesá s popsaným vztahem. V obecném případě je tento tlumící koeficient roven $\frac{h}{m} = 2 \delta $. Protože výkon je práce provedená za čas, tak výkon vztažený na jednu periodu je právě roven vztahu v rámečku. Maximum opět nastává, když derivace výkonu podle úhlové frekvence budící síly je roven 0.
