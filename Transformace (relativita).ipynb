# Michelson-Morleyův experiment


Paprsek vyrazí z počátku, rozpojí se v bodě B, aby se pak srazili a interferovali. Protože by se soustava měla vůči předpokládanému éteru posouvat nějakou rychlostí u, bude čas potřebný pro cestu z bodu B do E rovna:
$$
t = t_{1} + t_{2}
$$

$$
ct_{1} = L+ ut_{1}
$$

$$
ct_{2} = L- ut_{2}
$$
$$
t = \frac{2Lc}{c^{2}-v^{2}}
$$
Pro paprsek letící k C platí, že při letu do bodu C opíše přeponu rovnoramenného trojúhelníku a po cestě zpět urazí stejnou vzdálennost. Po cestě směrem k C však zrcadélko uletí o vzdálennost $ut_{3}$, protože se též posunuje. Platí tedy, že:
$$
t_{3} = \frac{2L}{\sqrt{c^{2}-u^{2}}}
$$
*Proč se to pootočením o 90° spraví??*
Když upravíme  u vztahu pro výpočet času z B do E pomocí Lorentzova vztahu - a sice budeme počítat s kontrakcí délky, tak zjišťujeme, že:
$$
t_{1} + t_{2} = t_{3}
$$
Což vysvětluje to, že nedochází k žádnému posunu. Krásný příklad na Lorentzovu transformaci jsou $\mu$ miony.
