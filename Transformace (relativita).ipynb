# Michelson-Morleyv experiment


Paprsek vyraz z potku, rozpoj se v bod B, aby se pak srazili a interferovali. Protoe by se soustava mla vi pedpokldanmu teru posouvat njakou rychlost u, bude as potebn pro cestu z bodu B do E rovna:
$$
t = t_{1} + t_{2}
$$

$$
ct_{1} = L+ ut_{1}
$$

$$
ct_{2} = L- ut_{2}
$$
$$
t = \frac{2Lc}{c^{2}-v^{2}}
$$
Pro paprsek letc k C plat, e pi letu do bodu C ope peponu rovnoramennho trojhelnku a po cest zpt uraz stejnou vzdlennost. Po cest smrem k C vak zrcadlko ulet o vzdlennost $ut_{3}$, protoe se t posunuje. Plat tedy, e:
$$
t_{3} = \frac{2L}{\sqrt{c^{2}-u^{2}}}
$$
*Pro se to pootoenm o 90 sprav??*
Kdy upravme  u vztahu pro vpoet asu z B do E pomoc Lorentzova vztahu - a sice budeme potat s kontrakc dlky, tak zjiujeme, e:
$$
t_{1} + t_{2} = t_{3}
$$
Co vysvtluje to, e nedochz k dnmu posunu. Krsn pklad na Lorentzovu transformaci jsou $\mu$ miony.
