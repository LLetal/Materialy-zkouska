# Steinerova vta
Mme-li dv rovnobn osy, tak momenty setrvanosti vi tmto dvou osm jsou ve vztahu:

$$
J = J_{T} + MR_{T}^{2}
$$

## Moment setrvanosti u kolmch os
zjistit moment setrvanosti vzhledem k njak ose, na kterou jsou kolm dv dal, plat vztah:
$$
J_{z} = J_{x} + J_{y}
$$
## Chaslesv teorm
Libovoln pohyb tuhho tlesa lze sloit z rotace kolem osy a translace podl tto osy
$$
v = v_{R} + v_{T}
$$

## Moment setrvanosti vzhledem k obecn ose
Vyuvme tzv. tenzor setrvanosti.

$$
\begin{pmatrix}
J_{xx} & D_{xy} & D_{xz}\\
D_{xy} & J_{yy} & D_{yz}\\
D_{xz} & D_{yz} & J_{zz}
\end{pmatrix}
$$


Prvn len na diagonle pedstavuje vzdlennost od osy x, druh od osy y, tet od osy z. Zbyl leny od pslunch ploch(nap $D_{xy}$ je vzdlenost od roviny xy).

## Setrvank



Kdy pi translanm pohybu na tleso s njakou hybnost psobme silou F, tak se za touto silou zane pohybovat. U setrvanku pouijeme pslun veliiny (Moment sly a moment hybnosti), v takovm ppad t bude tleso otet do smru momentu sly.

```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
from scipy.integrate import dblquad
from scipy.integrate import tplquad
import pandas as pd
```

```python
def f(angle): # funkce polomru toc ki v zvislosti na hlu
    return(1-np.pi/4 +angle**(1/10))
```

```python
angles=  np.linspace(np.pi/4,np.pi/2,1000)
r = f(angles)
x_souradnice = np.insert(r*np.cos(angles),0,0)
y_souradnice = np.insert(r*np.sin(angles),0,0)

plt.plot(x_souradnice,y_souradnice)
plt.grid()
plt.ylim(-1.5,1.5)
plt.xlim(-1.5,1.5)
```
