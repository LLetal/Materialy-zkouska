# Gravitační pole - detailněji a složitěji

Okolo každého tělesa je Gravitační pole. Vlastnosti pole charakterizujeme dvěma základními parametry, a sice Intenzitou a Potenciálem.
$$
\vec{K} = -G \frac{M}{|\vec{r}|^{2}} \frac{\vec{r}}{|\vec{r}|}
$$
Intenzita gravitačního pole udává sílu, jíž by pole působilo v daném místě na těleso o jednotkové hmotnosti. 

Myšlenka potenciálu vychází z toho, že při přemístění tělesa v GP vykonáváme práci úměrnou:
$$
W_{AB} = m \int_{r_{A}}^{r_{B}}\vec{K} d\vec{r} = - \Delta E_{p}
$$
(Funguje to, protože intenzita je vlastně síla na hmotnost a integruji-li ji přes polohový vektor, získávám vlastně práci, kterou musím vykonat, abych přesunul těleso o jednotkové hmotnosti z jednoho bodu do druhého). U konzervativního pole je určena jen počáteční a koncovou polohou a nezávisí na trajektorii.
$$
-\phi(\vec{r}) = \frac{E_{p}(\vec{r})}{m}
$$
$\phi$ definujeme jako práci, kterou musí vykoná těleso jednotkové hmotnosti z místa nulového potenciálo do nějakého bodu
$$
E_{p} = -W_{\infinity,r_{A}} = -\int_{\inf}^{r_{A}}\vec{F}d\vec{r} = - GMm \frac{1}{r_{A}}
$$
$$
\phi = -G\frac{M}{r}
$$
Při výpočtu potenciální energie E v tíhovém poli volíme jako nulový bod potenciálu povrch Země. Počítáme s homogenitou pole a vycházíme ze vztahu $mgh$, pročež je závislost lineární. Skutečný $E_p$ se na větších vzdálenostech významně liší. 

### Konzervatičnost gravitačního pole
Nezáleží na trajektorii při pohybu uvnitř pole. 


Prvotní intuitivní myšlenka: Když se pohybuju uvnitř pole, tak při je jedno jakou cestu zvolím, protože při ní budu muset překonat stejný počet vrstevnic potenciálu.

# Kosmické rychlosti

### První kosmická rychlost
$$
v = \sqrt{gR_{z}} = 7.9
$$
Minimální rychlost pro to, aby těleso nespadlo na povrch. Odvození z odstředivé síly.
$$
v > \sqrt{G \frac{M_{z}}{R_{z}}}
$$
Eliptická oběžná dráha

### Druhá kosmická rychlost (úniková)

Minimální rychlost, kterou těleso musí mít, aby uniklo jejímu gravitačního působení. Pohybuje se po parabolické trajektorii-je-li rychlost větším jedná se o hyperbolickou trajektorii.
$$
v_p = \sqrt{\frac{2GM}{r}}
$$

Počítáme-li s tím, že veškerou kinetickou energii těleso putující z gravitačního pole spotřebuje na jeho opuštění a s tím, že potenciální energie je po úniku nulová, musí notně platit, že:
$$
\frac{1}{2} m v^{2} - G\frac{mM}{r} = 0
$$
Odtud:
$$
v = \sqrt{\frac{2GM}{r}}
$$
