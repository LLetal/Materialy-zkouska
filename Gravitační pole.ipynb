# Gravitan pole - detailnji a sloitji

Okolo kadho tlesa je Gravitan pole. Vlastnosti pole charakterizujeme dvma zkladnmi parametry, a sice Intenzitou a Potencilem.
$$
\vec{K} = -G \frac{M}{|\vec{r}|^{2}} \frac{\vec{r}}{|\vec{r}|}
$$
Intenzita gravitanho pole udv slu, j by pole psobilo v danm mst na tleso o jednotkov hmotnosti. 

Mylenka potencilu vychz z toho, e pi pemstn tlesa v GP vykonvme prci mrnou:
$$
W_{AB} = m \int_{r_{A}}^{r_{B}}\vec{K} d\vec{r} = - \Delta E_{p}
$$
(Funguje to, protoe intenzita je vlastn sla na hmotnost a integruji-li ji pes polohov vektor, zskvm vlastn prci, kterou musm vykonat, abych pesunul tleso o jednotkov hmotnosti z jednoho bodu do druhho). U konzervativnho pole je urena jen poten a koncovou polohou a nezvis na trajektorii.
$$
-\phi(\vec{r}) = \frac{E_{p}(\vec{r})}{m}
$$
$\phi$ definujeme jako prci, kterou mus vykon tleso jednotkov hmotnosti z msta nulovho potencilo do njakho bodu
$$
E_{p} = -W_{\infinity,r_{A}} = -\int_{\inf}^{r_{A}}\vec{F}d\vec{r} = - GMm \frac{1}{r_{A}}
$$
$$
\phi = -G\frac{M}{r}
$$
Pi vpotu potenciln energie E v thovm poli volme jako nulov bod potencilu povrch Zem. Potme s homogenitou pole a vychzme ze vztahu $mgh$, proe je zvislost linern. Skuten $E_p$ se na vtch vzdlenostech vznamn li.

### Druh kosmick rychlost (nikov)

Minimln rychlost, kterou tleso mus mt, aby uniklo jejmu gravitanho psoben. Pohybuje se po parabolick trajektorii-je-li rychlost vtm jedn se o hyperbolickou trajektorii.
$$
v_p = \sqrt{\frac{2GM}{r}}
$$

Potme-li s tm, e vekerou kinetickou energii tleso putujc z gravitanho pole spotebuje na jeho oputn a s tm, e potenciln energie je po niku nulov, mus notn platit, e:
$$
\frac{1}{2} m v^{2} - G\frac{mM}{r} = 0
$$
Odtud:
$$
v = \sqrt{\frac{2GM}{r}}
$$
