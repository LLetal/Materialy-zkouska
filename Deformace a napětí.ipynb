## Hookv zkon
Kdy tahm za ty, tak se prodliuje ve smru podl tahn. Ve smru pnm se ale bude tak deformovat (ka se bude sniovat nebo zvyovat). Ty z 'normlnho' materilu se z  kdy j nathneme. Ale existuj materily, kde to je obrcen. Tlame-li tedy na ty ve smru x a svou dlku m rovn v tomto smru, tak je jeho deformace v tomto smru rovna $\epsilon_{xx} = \sigma_{xx} \frac{1}{E}$. Deformace ve smru y je rovna $\epsilon_{yy}= -\sigma_{xx} \frac{1}{k_t}= -\tau \epsilon_{xx}$. V rovnicch $k_l$ a $k_t$ pedstavuj pomrnou deformaci ve smru l a ve smru prezu. $\ni$ je potom jejich pomrem. Kladn Poissonv pomr odpovd 'normlnm materilm', kter se z kdy je nathneme. Zporn Poissonv pomr odpovd opaku. Poissonv pomr vdycky mus leet na intervalu $(-1,0.5)$. Spodn limit (-1) je dn tm, e materil s  im Poissonovm pomrem by se rozioval rychleji ne prodluoval, kdy bychom za nj tahali. Horn limit (0.5) si lze odvodit, pokud si pedstavme teba kostlu pod vodou. Na kadou stnu kostky psob stejn tlak. Lze si odvodit, e pokud $\ni =0.5$, tak se kostla mebude vbec deformovat a pro $\ni>0.5$ se bude zvtovat (co nedv smysl).
