# Deformace

Objemové síly - působí na všechny částice tělesa
Plošné síly - působí pouze na povrch tělesa a způsobují tedy deformaci v příslušném místě.
Pro popis míry deformace zavádíme veličinu napětí, která má tvar:
$$
\vec{\sigma} = \frac{d\vec{F}}{dS}

## Tenzor napětí
Napětí působící na obecné těleso popisujeme pomocí tenzoru napětí, který obsahuje informaci o napětí ve všech směrech. 
.
Je v něm uložená informace o všech napětích uvnitř tělesa. Napětí v nějakém směru tedy spočteme jako součin tenzoru a jednotkového vektoru zleva ve směru, který nás zajímá. Například pokud bychom chtěli získat informaci o napětí ve směru x, násobili bychom vektorem $\vec{n} = [1,0,0)$  
## Tenzor deformací
Stejně jako u napětí popisujeme deformaci tělesa pomocí tenzoru, který obsahuje všechny důležité informace.



Podobně jako u tenzoru napětí můžeme tenzor deformací zjistit, co se stane jednotlivými body tělesa. Pokud polohový vektor nějakého bodu v těleso přenásobíme tenzorem deformací získame polohu bodu po deformaci. Tenzor deformace lze získat buď ze známých deformací anebo z tenzoru napětí a materiálových konstant (viz. zobecněný Hookúv zákon).
## Hookův zákon
Když tahám za tyč, tak se prodližuje ve směru podél tahání. Ve směru příčném se ale bude také deformovat (šířka se bude snižovat nebo zvyšovat). Tyč z 'normálního' materiálu se zůží  když jí natáhneme. Ale existují materiály, kde to je obráceně. Tlačíme-li tedy na tyč ve směru x a svou délku má rovněž v tomto směru, tak je jeho deformace v tomto směru rovna $\epsilon_{xx} = \sigma_{xx} \frac{1}{E}$. Deformace ve směru y je rovna $\epsilon_{yy}= -\sigma_{xx} \frac{1}{k_t}= -\tau \epsilon_{xx}$. V rovnicích $k_l$ a $k_t$ představují poměrnou deformaci ve směru l a ve směru průřezu. $\ni$ je potom jejich poměrem. Kladný Poissonův poměr odpovídá 'normálním materiálům', které se zůží kdyť je natáhneme. Záporný Poissonův poměr odpovídá opaku. Poissonův poměr vždycky musí ležet na intervalu $(-1,0.5)$. Spodní limit (-1) je dán tím, že materiál s  ižším Poissonovým poměrem by se rozšiřoval rychleji než prodlužoval, když bychom za něj tahali. Horní limit (0.5) si lze odvodit, pokud si představíme třeba kostlu pod vodou. Na každou stěnu kostky působí stejný tlak. Lze si odvodit, že pokud $\ni =0.5$, tak se kostla mebude vůbec deformovat a pro $\ni>0.5$ se bude zvětšovat (což nedává smysl).
